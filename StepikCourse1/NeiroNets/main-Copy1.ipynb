```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
# RandomForestClassifier(random_state=0)
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/training_mush.csv')
```

```python
 X_test = pd.read_csv(r'testing_mush.csv')
```

```python
X_train = df.drop('class', axis=1)
y_train = df['class']
```

```python
rf_clf = RandomForestClassifier(random_state=0)


params = {
    'n_estimators':range(10,51,10),
'max_depth':range(1,13,2),
'min_samples_leaf':range(1,8),
'min_samples_split':range(2,10,2)   
}

grid_search_cv_clf = GridSearchCV(rf_clf, params, cv=3, n_jobs=-1)


grid_search_cv_clf.fit(X_train, y_train)


t = grid_search_cv_clf.best_params_
```

```python
rf_clf = RandomForestClassifier(random_state=0, max_depth=9,\
     min_samples_leaf=1,\
     min_samples_split=2,\
     n_estimators=10)\

rf_clf.fit(X_train, y_train)
imp = pd.DataFrame(rf_clf.feature_importances_, index=X_train.columns, columns=['importance'])
imp.sort_values('importance').plot(kind='barh', figsize=(12, 8))
plt.show()
```

```python
grid_search_cv_clf.best_params_
best_clf = grid_search_cv_clf.best_estimator_


result = best_clf.predict(X_test)
pd.Series(result).value_counts()

# best_clf = grid_search_cv_clf.best_params_

# Замени на две следующих

# grid_search_cv_clf.best_params_

# best_clf = grid_search_cv_clf.best_estimator_

# И в съедобности сделай так:
# result = best_clf.predict(X_test)
#  pd.Series(result).value_counts()
```
